```python
# creation de deuxieme couche caché
nombre_de_deuxiem_neud_cache = 6
classifier.add(Dense(units=nombre_de_deuxiem_neud_cache, activation="relu", kernel_initializer="uniform" ))
```

```python
classifier.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
classifier.fit(X_train,y_train,batch_size=10,epochs=100)
```

```python
#Prediction 
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

# Evaluation, Amelioration et Ajustement du ANN

```python
# Classification template
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# Importing the dataset
dataset = pd.read_csv('data/Churn_Modelling.csv')
X = dataset.iloc[:, 3:13].values
y = dataset.iloc[:, 13].values
# Encoding categorical data
# Encoding the Independent Variable
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X_1 = LabelEncoder()
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])
labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])

onehotencoder = OneHotEncoder(categorical_features = [1])
X = onehotencoder.fit_transform(X).toarray()
X = X[:,1:]
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
def build_classifier():
    # creation de première couche
    classifier = Sequential()
    entree = 11
    sortie = 1
    nombre_neud_cache = 6
    nombre_de_deuxiem_neud_cache = 6
    classifier.add(Dense(units=nombre_neud_cache, activation="relu", 
                         kernel_initializer="uniform",input_dim=entree))
    classifier.add(Dense(units=nombre_de_deuxiem_neud_cache, activation="relu", 
                         kernel_initializer="uniform" ))
    classifier.add(Dense(units=sortie, activation="sigmoid",
                         kernel_initializer="uniform" ))
    classifier.compile(optimizer="adam",loss="binary_crossentropy",metrics=['accuracy'])
    classifier.summary()
    return classifier
```

```python
#build_classifier().summary()
cross_classifier = KerasClassifier(build_fn=build_classifier,batch_size=10,epochs=100)
```

```python
precision = cross_val_score(estimator = cross_classifier,X=X_train,y=y_train,cv=10)
```

```python
print precision.mean()
print precision.std()*precision.std()
```

# gestion de Sur-apprentissage avec Dropout

```python
from keras.layers import Dropout
```

```python
# creation de première couche
classifier = Sequential()
nombre_de_premier_neud_cache = 6
entree = 11
classifier.add(Dense(units=nombre_de_premier_neud_cache, 
                     activation="relu", 
                     kernel_initializer="uniform",
                     input_dim=entree ))
classifier.add(Dropout(rate=0.1))
# creation de deuxieme couche caché
nombre_de_deuxiem_neud_cache = 6
classifier.add(Dense(units=nombre_de_deuxiem_neud_cache,
                     activation="relu",
                     kernel_initializer="uniform" ))
classifier.add(Dropout(rate=0.1))
#sortie
sortie = 1
classifier.add(Dense(units=sortie, 
                     activation="sigmoid",
                     kernel_initializer="uniform" ))
classifier.compile(optimizer="adam",
                   loss="binary_crossentropy",
                   metrics=["accuracy"])
classifier.fit(X_train,y_train,batch_size=10,epochs=100)
```

```python
#Prediction 
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

# Optimastion avec great seach

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
```

```python
def build_classifier(optimizer):
    # creation de première couche
    classifier = Sequential()
    entree = 11
    sortie = 1
    nombre_neud_cache = 6
    nombre_de_deuxiem_neud_cache = 6
    classifier.add(Dense(units=nombre_neud_cache, activation="relu", 
                         kernel_initializer="uniform",input_dim=entree))
    classifier.add(Dense(units=nombre_de_deuxiem_neud_cache, activation="relu", 
                         kernel_initializer="uniform" ))
    classifier.add(Dense(units=sortie, activation="sigmoid",
                         kernel_initializer="uniform" ))
    classifier.compile(optimizer=optimizer,
                       loss="binary_crossentropy",
                       metrics=['accuracy'])
    classifier.summary()
    return classifier
```

```python
#Nous allons essayer d'optimiser batch_size et epochs
classifier = KerasClassifier(build_fn=build_classifier,batch_size=10,epochs=100)
```

```python
parameters = {"batch_size": [25,32,50],"epochs": [100,200,500,700],"optimizer":["adam","rmsprop"]}
grid_search = GridSearchCV(estimator=classifier,
                           param_grid=parameters,
                           scoring="accuracy",
                           cv=10)
grid_search.fit(X_train,y_train)
```
